```python
import pandas as pds
import numpy as np
import matplotlib.pyplot as plt
import collections
import seaborn as sns

ds = pds.read_csv("stats_annuelles.csv", delimiter=";") #Chargement du dataset
ds
```

```python
for c in ds.drop(["Annee"], axis=1).columns:
    fig = plt.figure(figsize=(15,5))
    ax1 = fig.add_subplot(111)
    ax1.plot(ds["Annee"], ds[c]) 
    ax1.set_title("- L'évolution des "+c+" dans le département des Landes -")
    plt.show()
```
