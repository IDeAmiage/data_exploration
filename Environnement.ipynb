```python
# Température moyen par communes
# Température moyen sur un mois
# moyen d'humitié
# moyen vitesse du vent
# moyen couche de neige

communes_df = df.groupby('communes (name)')
mois_df = df.groupby('mois_de_l_annee')

temperature_df_moy = communes_df['Température (°C)'].mean()
humidite_df_moy = mois_df['Humidité'].mean()
vitesse_vent_df_moy = mois_df['Vitesse du vent moyen 10 mn'].mean()
couche_neige_df_moy = mois_df['Hauteur totale de la couche de neige, glace, autre au sol'].mean()

print(temperature_df_moy)
print(humidite_df_moy)
print(vitesse_vent_df_moy)
print(couche_neige_df_moy)
```
